```python
import os
import glob


def extract_from_csv(path):
    
    csv_pattern = os.path.join(path, '.csv')  # Builds "path/.csv" safely across OS types
    csv_files = glob.glob(csv_pattern)
    
    extracted_data_csv = pd.DataFrame(columns=['car_model', 'year_of_manufacture', 'price', 'fuel'])
    
    for file in csv_files:
        df = pd.read_csv(file)
        extracted_data_csv = pd.concat([extracted_data_csv, df], ignore_index=True)
    
    return extracted_data_csv
```
